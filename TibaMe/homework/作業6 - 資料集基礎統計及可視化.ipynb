### Read Total profit of all months and show it using a line plot
* X label name = Month Number
* Y label name = Total profit

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('company_sales_data.csv')
df.head()
```

```python
## write your code here
X = df['month_number']
Y = df['total_profit']

plt.plot(X, Y)
plt.title('Company profit per month')
plt.ylabel('Profit in dollar')
plt.xlabel('Month Number')
plt.xticks(np.linspace(1, 12, 12))
plt.yticks(np.linspace(100000, 500000, 5))
plt.show()
```

```python
## write your code here
X = df['month_number']
Y = df['toothpaste']

plt.scatter(X, Y, c='cornflowerblue', label='Tooth paste Sales data')
plt.rc('grid', linestyle="--", color='gray')
plt.ylabel('Number of units Sold')
plt.xlabel('Month Number')
plt.xticks(np.linspace(1, 12, 12))
plt.legend(loc='best')
plt.title('Tooth paste Sales data')
plt.grid()
plt.show()
```

### Calculate total sale data for last year for each product and show it using a Pie chart  
【Note】In Pie chart display Number of units sold per year for each product in percentage

```python
## write your code here
total_units = df['total_units'].sum()
labels = df.columns.drop(['total_profit', 'total_units', 'month_number'])

sizes = []
for col in labels:
    col_sum = df[col].sum()   
    sizes.append(round(col_sum / total_units * 100, 2))

plt.figure()
explode = (0, 0.3, 0, 0, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
plt.legend(loc="lower right")
plt.axis('equal')
plt.title('Sales data', pad=30)
plt.show()
```
